
Plotting simple quantities of a pandas dataframe
=================================================

This example loads from a CSV file data with mixed numerical and
categorical entries, and plots a few quantities, separately for females
and males, thanks to the pandas integrated plotting tool (that uses
matplotlib behind the scene).

See http://pandas.pydata.org/pandas-docs/stable/visualization.html



```python
import pandas

# Load the data
df = pandas.read_csv('./brain_size.csv', sep=';', na_values=".")
del df["Unnamed: 0"]


# Scatter matrices for different columns


```

```python
df.head()
```

df.describe()

```python
import matplotlib.pyplot as plt
# Box plots of different columns for each gender
#'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count'
df.boxplot(column = ['FSIQ'], by = 'Gender')
plt.show()
```

```python
from pandas.tools import plotting
df_male = df[df['Gender']=='Male']
df_female = df[df['Gender']=='Female']
df_female_tall = df[(df['Gender']=='Female') & (df['Height']>=68)]
plotting.scatter_matrix(df_male[['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']], c='r')
plotting.scatter_matrix(df_female[['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']])
plotting.scatter_matrix(df_female_tall[['FSIQ', 'VIQ', 'PIQ', 'Weight', 'Height', 'MRI_Count']])
plt.show()
```
