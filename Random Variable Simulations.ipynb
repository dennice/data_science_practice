```python
% matplotlib inline
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import math
```

```python
# a normal random variable has to parameters : m and s2

#parameters
m = 0
s2 = 2
s=np.sqrt(s2)
#define domain values
X1=np.linspace(start=-10,stop=10,num=1000)
from scipy.stats import norm
# define the values for the normal random variables based on domain X1
Y1=norm.pdf(X1,m,s)
```

```python
plt.plot(X1,Y1)
```

```python
#look at the distribution with different values of mean and variance

X1=np.linspace(start=-10,stop=10,num=1000)
for m0 in range(-100,101):
    m=m0
    s2=abs(-100)
    s= np.sqrt(s2)
    Y1=norm.pdf(X1,m,s)
    plt.plot(X1,Y1)
```
